```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
```

```python
plt.plot(yield_apples);
```

```python
plt.figure(figsize=(10,6))
years = np.arange(2000, 2006,1)
years.shape, len(yield_apples)
plt.plot(years, yield_apples )
plt.title('Yield of apples')
plt.xlabel('Year')
plt.ylabel('Yield')
```

```python
years = range(2000, 2012, 1)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
```

```python
plt.figure(figsize=(10,6))
plt.plot(years, apples,marker = 'x' , color = 'red', linestyle='--', linewidth='2', ms='10', mec='green', mew ='3', mfc='yellow', alpha =1.6)
plt.plot(years, oranges, marker = 'o', mew = '6', alpha=.6)
plt.xlabel('year')
plt.ylabel('Yield per/hct')
plt.title('Yield');
```

```python
plt.figure(figsize=(10,6))
plt.plot(years, apples, 'x--g')
plt.plot(years, oranges, marker = 'o', mew = '6', alpha=.6)
plt.xlabel('year')
plt.ylabel('Yield per/hct')
plt.title('Yield');
```

```python
plt.figure(figsize=(10,6))
plt.plot(years, oranges, 'or')
plt.title('Yield');
```

```python
sns.__version__
```

```python
plt.rcParams['figure.figsize'] = (9,5)
plt.rcParams['font.size'] = 14
plt.rcParams['figure.facecolor'] = '#00000000' 
```

```python
sns.set_style('whitegrid')
```

```python
plt.figure(figsize=(9,5))
plt.plot(years, apples, 's-b')
plt.plot(years, oranges, marker='o', linestyle= '--', mec='red', mfc='red', mew='2', color = 'red')
plt.xlabel('Year')
plt.ylabel('Yeild per hect')
plt.title('Crop Yield')
plt.legend(['apples', 'oranges']);
```

```python
plt.scatter(flowers_df.sepal_length, flowers_df.sepal_width)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Dimensions');
```

```python
plt.figure(figsize=(10,6))
plt.title('Sepal Dimensions')
sns.scatterplot(x =flowers_df.sepal_length,y= flowers_df.sepal_width, hue = flowers_df.species , s =200);
```

```python
plt.figure(figsize=(10,6))
plt.title('Sepal Dimensions')
sns.scatterplot(x='sepal_length', y = 'sepal_width', data=flowers_df, hue='species', s=100);
```

```python
setosa_df = flowers_df[flowers_df.species == 'setosa']
versicolor_df = flowers_df[flowers_df.species == 'versicolor']
virginica_df = flowers_df[flowers_df.species == 'virginica']
```

```python
setosa_df.species.unique()
```

```python
plt.hist('sepal_length', data=setosa_df, color='b' ,bins=np.arange(4, 8, 0.25))
plt.hist('sepal_length', data=versicolor_df, color='g',bins=np.arange(4, 8, 0.25))
plt.hist('sepal_length', data=virginica_df, color='r',bins=np.arange(4, 8, 0.25))
```

```python
plt.hist(setosa_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));
plt.hist(versicolor_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));
```

```python
plt.title('Sepal Dimensions')
plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width,virginica_df.sepal_width], stacked=True, bins=np.arange(2,5,0.25))
plt.legend(['setosa', 'versicolor', 'virginica'])
```
